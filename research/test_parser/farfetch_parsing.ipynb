```python
import requests
from fake_useragent import UserAgent
from bs4 import BeautifulSoup
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from bs4 import BeautifulSoup
import codecs
import re
from webdriver_manager.chrome import ChromeDriverManager
import json
import pandas as pd
```

После импортов открываем через **selenium** google (можно просто включить **VPN** на устройстве), 
хотя можно заморочиться над бесплатными **proxy**.

```python
driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()))
```

```python
%%time
# Здесь можно добавить парсинг количетсва страниц, чтобы вручную их не задавать
for page in range(1, 216):
    val = f"https://www.farfetch.com/uk/shopping/men/trainers-2/items.aspx?page={page}&view=96&sort=3&scale=283"
    wait = WebDriverWait(driver, 5)
    driver.get(val)

    get_url = driver.current_url
    wait.until(EC.url_to_be(val))
    page_source = driver.page_source

    soup = BeautifulSoup(page_source, features="html.parser")
    data = [
        json.loads(x.string)
        for x in soup.find_all("script", type="application/ld+json")
    ]

    if page == 1:
        df = pd.DataFrame(data[0])
    else:
        df_1 = pd.DataFrame(data[0])
        df = pd.concat([df, df_1], ignore_index=True)
```

```python
# получаем данные из словаря
data = (
    pd.json_normalize(df.itemListElement)
    .drop_duplicates("offers.url")
    .reset_index(drop=True)
)
```

```python
# Нам нужно только первое фото
data["right_side_photo_image"] = [x[0] for x in data["image"]]
data["full_name"] = data["brand.name"] + " " + data["name"]
# Заменяем мусорный части в имени на те, которые указаны при на сайте
data["full_name"] = (
    data["full_name"].replace("&quot", "'", regex=True).replace(";", "", regex=True)
)
```

```python
data[
    [
        "brand.name",
        "name",
        "offers.price",
        "offers.priceCurrency",
        "offers.url",
        "right side photo_image",
        "full_name",
    ]
]
```

```python
# сохраняем весь df
data.to_csv("farfetch_parsing_full_df.csv", index=0)
```
